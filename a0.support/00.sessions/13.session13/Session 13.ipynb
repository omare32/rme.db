```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
url='http://books.toscrape.com/index.html'
soup=BeautifulSoup(requests.get(url).content,'html.parser')
soup
```

```python
domain=url[:url.rfind('/')+1]


for cat in soup.find('ul',attrs={'class':'nav nav-list'}).find('ul').find_all('a'):
    print(cat.get_text().strip())
    print(domain+cat.get('href'))
```
