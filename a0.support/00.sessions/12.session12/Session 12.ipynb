```python
url='https://ocw.mit.edu/courses/6-006-introduction-to-algorithms-spring-2020/resources/lecture-videos/'
response=requests.get(url)
soup=BeautifulSoup(response.text,'html.parser')
domain='https://ocw.mit.edu'
file_name=soup.find('a',attrs={'class':'text-capitalize m-0 text-white'}).get_text()+'.csv'
with open(file_name,'w',newline='',encoding='utf8') as f:
    header=['Lecture Name','Link']
    writer=csv.writer(f)
    writer.writerow(header)
    for lec in soup.find_all('a',attrs={'class':'resource-list-title'}):
        lec_name=lec.get_text()
        lec_link=domain+lec.get('href')
        writer.writerow([lec_name,lec_link])

  
```

```python
df=pd.read_csv(file_name)
```

```python
rates={'One':1,'Two':2,'Three':3,'Four':4,'Five':5}
url=f'http://books.toscrape.com/catalogue/category/books/mystery_3/page-1.html'
soup=BeautifulSoup(requests.get(url).text,'html.parser')
file_name=soup.find('div',attrs={'class':'page-header action'}).find('h1').get_text()
num_pages=int(soup.find('li',attrs={'class':'current'}).get_text().strip().split()[-1])

with open(file_name+'.csv','w',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(['Book Name','Price','rate'])
    for i in range(1,num_pages+1):
        url=f'http://books.toscrape.com/catalogue/category/books/mystery_3/page-{i}.html'
        soup=BeautifulSoup(requests.get(url).text,'html.parser')
        for article in soup.find_all('article',attrs={'class':'product_pod'}):
            book_name=article.find('h3').find('a').get('title')
            price=float(article.find('p',attrs={'class':'price_color'}).get_text()[2:])
            rate=rates[article.find('p',attrs={'class':'star-rating'}).get('class')[1]]
            writer.writerow([book_name,price,rate])
    
```

```python
df=pd.read_csv(file_name+'.csv')
```
