```python
df['year']=df.datetime.dt.year
df['day']=df.datetime.dt.day_name()
df['month']=df.datetime.dt.month
df
```

```python
def map_season(x):
    if x in [12,1,2]:
        return 'Winter'
    elif x in [3,4,5]:
        return 'Spring'
    elif x in [6,7,8]:
        return 'Summer'
    else:
        return 'Fall'
    
```

```python
df['season_2']=df.month.apply(map_season)
df
```

```python
def map_holiday_workingday(x):
    if str(x['holiday'])=='nan' and str(x['workingday'])!='nan':
        return 1 if x['workingday']==0 else 0
    else:
        return x['holiday']
    
df['holiday']=df.apply(map_holiday_workingday,axis=1)
df
```

```python
df.isnull().sum()
```

```python
df[df['workingday']==1].day.value_counts()
```

```python
def map_holi(x):
    if str(x['holiday'])=='nan' and str(x['workingday'])=='nan':
        if x['day']=='Saturday' or x['day']=='Sunday':
            return 1
        else:
            return 0
    else:
        return x['holiday']
    
df['holiday']=df.apply(map_holi,axis=1)
df
```

```python
df.isnull().sum()
```

```python
def last_map_holi(x):
    if x['holiday']==x['workingday']:
        if x['day']=='Saturday' or x['day']=='Sunday':
            return 1
        else:
            return 0
    else:
        return  x['holiday']
    
df['holiday']=df.apply(last_map_holi,axis=1)
df
```

```python
df.isnull().sum()
```

```python
df.drop(['season','workingday'],axis=1,inplace=True)
df
```

```python
def profit(x):
    revenue=(x['casual']*20) +(x['registered']*5)
    maint=1500/(365*24)
    profit=revenue-(revenue*0.14)-maint
    return profit
df['profit']=df.apply(profit,axis=1)
df
```
