```python
def map_bank(word):#OD,Overdraft,OVER DRAFT,Over Draft
    currencies=['EGP','SAR','USD','EUR']
    rt_currency=None
    rt_bank_name=None
    rt_od=False
    for currenc in currencies:
        if currenc in word:
            rt_currency=currenc
    for chk in ['OD','Overdraft','OVER DRAFT','Over Draft']:
        if chk in word:
            rt_od=True
        
    if 'HSBC' in word:
        rt_bank_name='HSBC'
    elif 'Ahly United bank in Bahrin' in word:
        rt_bank_name='Ahly United bank in Bahrin'
    elif 'Arab African International Bank' in word:
        rt_bank_name='Arab African International Bank'
    elif 'National Bank Of Eygpy' in word or 'NBE' in word:
        rt_bank_name='National Bank Of Eygpy'
    elif 'SUEZ Canal Bank' in word:
        rt_bank_name='SUEZ Canal Bank'
    elif 'Banque Misr' in word or 'Bank Misr' in word:
        rt_bank_name='Banque Misr'
    elif 'AAIB' in word:
        rt_bank_name='AAIB'
    elif 'EGYPTIAN Arab Land' in word:
        rt_bank_name='EGYPTIAN Arab Land'
    elif 'Commerical Internatoinal Bank' in word:
        rt_bank_name='CIB'
    elif 'Qatar National Bank Alahli' in word:
        rt_bank_name='QNB'
    elif 'Rowad Main Safe Box' in word:
        rt_bank_name='Rowad Main Safe Box'
    
    return pd.Series([rt_bank_name,rt_currency,rt_od])
    
```

```python
df[['bank_name','currency','over_draft']]=df.bank_account_name.apply(map_bank)
```
