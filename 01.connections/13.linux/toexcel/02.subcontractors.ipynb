```python
import mysql.connector as mysql
import pandas as pd
import re
import os

# Database connection details
DB_HOST = "10.10.11.242"
DB_USER = "omar2"
DB_PASSWORD = "Omar_54321"
DB_NAME = "RME_TEST"
TABLE_NAME = "RME_SubContractor_Report"

def sanitize_filename(name):
    """Sanitize project names for valid filenames."""
    name = name.strip().replace(" ", "_")  # Replace spaces with underscores
    name = re.sub(r'[\\/*?:"<>|]', "", name)  # Remove invalid characters
    name = re.sub(r'[^\x00-\x7F]+', "", name)  # Remove non-ASCII characters
    return name

try:
    # Establish the connection
    cnx = mysql.connect(
        host=DB_HOST,
        user=DB_USER,
        password=DB_PASSWORD,
        database=DB_NAME
    )
    
    if cnx.is_connected():
        print("Connection successful!")
        
        # Fetch data from the table
        query = f"SELECT * FROM {TABLE_NAME}"
        df = pd.read_sql(query, cnx)
        
        # Ensure the PROJECT column exists
        if "PROJECT" not in df.columns:
            print("Error: PROJECT column not found in the table.")
        else:
            # Save separate Excel files for each project
            for project, project_df in df.groupby("PROJECT"):
                safe_project_name = sanitize_filename(project)
                filename = f"{safe_project_name}.xlsx"
                
                # Ensure the filename is not empty after sanitization
                if not safe_project_name:
                    print(f"Skipping project '{project}' due to invalid filename.")
                    continue
                
                project_df.to_excel(filename, index=False)
                print(f"Saved: {filename}")

except mysql.Error as e:
    print(f"Error connecting to database: {e}")

finally:
    # Ensure connection is closed properly
    if 'cnx' in locals() and cnx.is_connected():
        cnx.close()
        print("Connection closed.")
```
