```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load data from Excel file (replace 'your_file.xlsx' with the actual file path)
file_path = 'ubari.xlsx'
df = pd.read_excel(file_path)
```

```python
# Extract data from the DataFrame
months = df['Month']
cash_in = df['Cash In']
cash_out = df['Cash Out']
```

```python
# Create a figure and axis
plt.figure(figsize=(10, 6))
plt.plot(months, cash_in, label='Cash In', marker='o', linestyle='-')
plt.plot(months, cash_out, label='Cash Out', marker='x', linestyle='-')
```

```python
# Customize the plot
plt.title('Monthly Cash In vs Cash Out')
plt.xlabel('Month')
plt.ylabel('Amount (USD)')
plt.grid(True)
plt.legend()
```

```python
# Set x-axis ticks to start from 1
plt.xticks(np.arange(1, len(months) + 1, step=1))
```

```python
# Show the plot
plt.tight_layout()
plt.show()
```
